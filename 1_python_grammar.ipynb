# if문

```python
male = True
if male:
  print('I am a man.')
else:
  print('I am not a man.')
```

```python
a = 5
b = 3
print(a < b)
print(a > b)
print(a == b)
print(a != b)
```

```python
if a > b:
  print('a is bigger than b.')
else:
  print('b is bigger than a.')
```

```python
a = True
b = False
print(a and b)
print(a or b)
print(not a)
```

```python
a = [1, 2, 3, 4, 5]
print(3 in a)
print(0 in a)
```

# for문

```python
a = ['one', 'two', 'three']
for i in a:
  print(i)
```

```python
for i in a:
  if i.count('t') > 0:
    print(i, 'has t.')
```

```python
result = 0
for i in range(10):
  result = result + i
print(result)

print([i * 10 for i in range(10)])
```

# 함수

```python
def add(a, b):
  return a + b

print(add(3, 4))
```

```python
def for_print(num, string):
  for i in range(num):
    print(string)

for_print(3, 'Hello Python')
```
