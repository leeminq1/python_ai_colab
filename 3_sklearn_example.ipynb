# 예제 1. Digit Classification

```python
from sklearn import datasets

digits = datasets.load_digits()

print(digits.images.shape)
print(digits.target.shape)
print(digits.images[0])
print(digits.target[0])
```

```python
import matplotlib.pyplot as plt

_, axes = plt.subplots(nrows=1, ncols=5, figsize=(10, 5))
for ax, image, target in zip(axes, digits.images, digits.target):
    ax.set_axis_off()
    ax.imshow(image, cmap='gray')
    ax.set_title('Target: ' + str(target))
```

```python
num_images = len(digits.images)
images = digits.images.reshape(num_images, -1)

print(images.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(images, digits.target, test_size=0.2, shuffle=False)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression

classifier = LogisticRegression(random_state=0)

classifier.fit(X_train, y_train)

print(classifier.score(X_test, y_test))
```

```python
y_test_pred = classifier.predict(X_test)

_, axes = plt.subplots(nrows=1, ncols=5, figsize=(10, 5))
for ax, image, pred in zip(axes, X_test, y_test_pred):
    ax.set_axis_off()
    image = image.reshape(8, 8)
    ax.imshow(image, cmap='gray')
    ax.set_title('Prediction: ' + str(pred))
```

```python
from sklearn import metrics

print(metrics.classification_report(y_test, y_test_pred))
```

```python
from sklearn.metrics import ConfusionMatrixDisplay

ConfusionMatrixDisplay.from_predictions(y_test, y_test_pred)
plt.show()
```

# 예제 4. Underfitting and Overfitting

```python
import numpy as np

X = np.linspace(0, 10)
y = np.sin(X)

X = X.reshape(-1, 1)

print(X.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
import matplotlib.pyplot as plt

plt.scatter(X_train, y_train)
plt.scatter(X_test, y_test)
plt.show()
```

```python
from sklearn import linear_model

regressor = linear_model.LinearRegression()

regressor.fit(X_train, y_train)

print(regressor.score(X_train, y_train))
print(regressor.score(X_test, y_test))
```

```python
y_pred = regressor.predict(X)

plt.scatter(X_train, y_train)
plt.scatter(X_test, y_test)
plt.plot(X, y_pred)
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures

polynomial = PolynomialFeatures(10)

X_train_ = polynomial.fit_transform(X_train)

X_ = polynomial.transform(X)
X_test_ = polynomial.transform(X_test)

print(X_.shape)
print(X_train_.shape)
print(X_test_.shape)
```

```python
regressor = linear_model.LinearRegression()

regressor.fit(X_train_, y_train)

print(regressor.score(X_train_, y_train))
print(regressor.score(X_test_, y_test))
```

```python
y_pred = regressor.predict(X_)

plt.scatter(X_train, y_train)
plt.scatter(X_test, y_test)
plt.plot(X, y_pred)
plt.show()
```

```python
polynomial = PolynomialFeatures(20)

X_train_ = polynomial.fit_transform(X_train)

X_ = polynomial.transform(X)
X_test_ = polynomial.transform(X_test)

print(X_.shape)
print(X_train_.shape)
print(X_test_.shape)
```

```python
regressor = linear_model.LinearRegression()

regressor.fit(X_train_, y_train)

print(regressor.score(X_train_, y_train))
print(regressor.score(X_test_, y_test))
```

```python
y_pred = regressor.predict(X_)

plt.scatter(X_train, y_train)
plt.scatter(X_test, y_test)
plt.plot(X, y_pred)
plt.show()
```

# 예제 6. Preprocessing

```python
from sklearn import datasets

X, y = datasets.load_digits(return_X_y=True)

print(X.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_ = scaler.fit_transform(X_train)
X_test_ = scaler.transform(X_test)

print(X_train[0])
print(X_train_[0])
```

```python
from sklearn.linear_model import RidgeClassifier

classifier = RidgeClassifier()

classifier.fit(X_train_, y_train)

print(classifier.score(X_test_, y_test))
```

```python
from sklearn.metrics import ConfusionMatrixDisplay

y_test_pred = classifier.predict(X_test_)

ConfusionMatrixDisplay.from_predictions(y_test, y_test_pred)
plt.show()
```

# 예제 7. Hyperparameter Search

```python
from sklearn import datasets

X, y = datasets.load_iris(return_X_y=True)

print(X.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV

classifier = KNeighborsClassifier()

param_grid = {
    'n_neighbors': [3, 5, 10],
    'p': [1, 2]}
search = GridSearchCV(classifier, param_grid)

search.fit(X_train, y_train)

print(search.score(X_test, y_test))
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
import matplotlib.pyplot as plt

y_test_pred = search.predict(X_test)

ConfusionMatrixDisplay.from_predictions(y_test, y_test_pred)
plt.show()
```

```python
import pandas as pd

results = pd.DataFrame(search.cv_results_)

results
```
