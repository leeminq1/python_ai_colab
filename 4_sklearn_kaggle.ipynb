# Competition 1. Diabetes Prediction

건강과 관련된 상태 데이터를 이용하여 당뇨병 여부를 예측하는 문제이다.

### INPUT  
- **HighBP**: 0 = no high BP 1 = high BP
- **HighChol**: 0 = no high cholesterol 1 = high cholesterol
- **CholCheck**: 0 = no cholesterol check in 5 years 1 = yes cholesterol check in 5 years
- **BMI**: Body Mass Index
- **Smoker**: Have you smoked at least 100 cigarettes in your entire life? [Note: 5 packs = 100 cigarettes] 0 = no 1 = yes
- **Stroke**: (Ever told) you had a stroke. 0 = no 1 = yes
- **HeartDiseaseorAttack**: coronary heart disease (CHD) or myocardial infarction (MI) 0 = no 1 = yes
- **PhysActivity**: physical activity in past 30 days - not including job 0 = no 1 = yes
- **Fruits**: Consume Fruit 1 or more times per day 0 = no 1 = yes
- **Veggies**: Consume Vegetables 1 or more times per day 0 = no 1 = yes
- **HvyAlcoholConsump** (adult men >=14 drinks per week and adult women>=7 drinks per week) 0 = no 1 = yes
- **AnyHealthcare**: Have any kind of health care coverage, including health insurance, prepaid plans such as HMO, etc. 0 = no 1 = yes
- **NoDocbcCost**: Was there a time in the past 12 months when you needed to see a doctor but could not because of cost? 0 = no 1 = yes
- **GenHlth**: Would you say that in general your health is: scale 1-5 1 = excellent 2 = very good 3 = good 4 = fair 5 = poor
- **MentHlth**: days of poor mental health scale 1-30 days
- **PhysHlth**: physical illness or injury days in past 30 days scale 1-30
- **DiffWalk**: Do you have serious difficulty walking or climbing stairs? 0 = no 1 = yes
- **Sex**: 0 = female 1 = male
- **Age**: 13-level age category (_AGEG5YR see codebook) 1 = 18-24 9 = 60-64 13 = 80 or older
- **Education**: Education level (EDUCA see codebook) scale 1-6 1 = Never attended school or only kindergarten 2 = elementary etc.
- **Income**: Income scale (INCOME2 see codebook) scale 1-8 1 = less than $10,000 5 = less than $35,000 8 = $75,000 or more

### OUTPUT  
- **Diabetes_binary**: 0 = no diabetes 1 = prediabetes 2 = diabetes

```python
import pandas as pd

df = pd.read_csv('diabetes_train.csv', index_col=0)
df.drop(["Education" ,"GenHlth"], axis=1, inplace=True)

df
```

```python
df.isnull().sum()
```

```python
X_train=df.iloc[:,1:]
y_train=df.iloc[:,0]

print(X_train.shape)
print(y_train.shape)
```

```python
X_test_real = pd.read_csv('diabetes_X_test.csv', index_col=0)
X_test_real.drop(["Education","GenHlth" ], axis=1, inplace=True)
X_test_real
```

```python
# 평가지표 출력하는 함수 설정
def get_clf_eval(y_test, y_pred):
    confusion = confusion_matrix(y_test, y_pred)
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    F1 = f1_score(y_test, y_pred)
    AUC = roc_auc_score(y_test, y_pred)
    
    print('오차행렬:\n', confusion)
    print('\n정확도: {:.4f}'.format(accuracy))
    print('정밀도: {:.4f}'.format(precision))
    print('재현율: {:.4f}'.format(recall))
    print('F1: {:.4f}'.format(F1))
    print('AUC: {:.4f}'.format(AUC))
    

# 평가지표를 조사하기 위한 새로운 함수 생성
def get_eval_by_threshold(y_test, pred_proba_c1, thresholds):
    #thresholds list 객체 내의 값을 iteration 하면서 평가 수행
    for custom_threshold in thresholds:
        binarizer = Binarizer(threshold=custom_threshold).fit(pred_proba_c1)
        custom_predict = binarizer.transform(pred_proba_c1)
        print('\n임계값: ', custom_threshold)
        get_clf_eval(y_test, custom_predict)
```

```python
# split 
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, roc_auc_score
from sklearn.metrics import f1_score, confusion_matrix, precision_recall_curve, roc_curve
from sklearn.preprocessing import StandardScaler, Binarizer
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

# 모델 변경 및 grid search 적용
# penalty = ['l2'] # 페널티(penalty) 하이퍼파라미터 값의 후보를 만듭니다.
# C = np.logspace(0, 4, 10) # 규제 하이퍼파라미터 값의 후보 범위를 만듭니다.
# hyperparameters = dict(C=C, penalty=penalty) # 하이퍼파라미터 후보 딕셔너리를 만듭니다.

# lr= LogisticRegression(random_state=0,max_iter=1000)
# lr_clt = GridSearchCV(lr, hyperparameters, cv = 5,verbose=0)
params = { 'max_depth' : [6, 8, 10, 12],
           'min_samples_split' : [16, 24]
            }

de_clt = DecisionTreeClassifier()
lr_clt = GridSearchCV(de_clt, param_grid=params, cv = 5,scoring='accuracy',verbose=1)

# StandardScaler 클래스를 상용하여 데이터 세트에 스케일링 적용
# 기존에 가지고 있는 테스트 set
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_train)

# test samle이 없으므로 임의로 나눔
# 기존 X_sclaed 값을 --> X_train / X_text로 나눔
X_scaled_train, X_scaled_test, y_sacled_train, y_scaled_test = train_test_split(X_scaled, y_train, test_size = 0.1)

# 학습해서 결과 계속 확인할 것 
lr_clt.fit(X_scaled_train, y_sacled_train)
pred = lr_clt.predict(X_scaled_test)
get_clf_eval(y_scaled_test, pred)

print("---------------")

# 임계값 변경
# thresholds = [0.3, 0.33, 0.36, 0.39, 0.42, 0.45, 0.48, 0.50]
# pred_proba = lr_clt.predict_proba(X_scaled_test)
# get_eval_by_threshold(y_scaled_test, pred_proba[:, 1].reshape(-1, 1), thresholds)
```

```python
# 나중에 결과 예측할때 사용 
X_scaled_real=scaler.transform(X_test_real)
y_test_pred=lr_clt.predict(X_scaled_real)
```

```python
import numpy as np

y_test_pred = pd.DataFrame(y_test_pred.reshape(-1, 1), index=X_test_real.index, columns=['Diabetes_binary'])

y_test_pred
```

# Competition 2. Wine Quality Prediction

### OUTPUT

```python
import pandas as pd

df = pd.read_csv('wine_train.csv', index_col=0)
df.drop(["sulphates" ], axis=1, inplace=True)

df
```

```python
df.isnull().sum()
```

```python
X_train=df.iloc[:,:10]
y_train=df.iloc[:,10]

print(X_train.shape)
print(y_train.shape)
```

```python
X_test_real = pd.read_csv('wine_X_test.csv', index_col=0)
X_test_real.drop(["sulphates" ], axis=1, inplace=True)

X_test_real
```

```python
# 평가지표 출력하는 함수 설정
def get_clf_eval(y_test, y_pred):
    confusion = confusion_matrix(y_test, y_pred)
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    F1 = f1_score(y_test, y_pred)
    AUC = roc_auc_score(y_test, y_pred)
    
    print('오차행렬:\n', confusion)
    print('\n정확도: {:.4f}'.format(accuracy))
    print('정밀도: {:.4f}'.format(precision))
    print('재현율: {:.4f}'.format(recall))
    print('F1: {:.4f}'.format(F1))
    print('AUC: {:.4f}'.format(AUC))
    

# 평가지표를 조사하기 위한 새로운 함수 생성
def get_eval_by_threshold(y_test, pred_proba_c1, thresholds):
    #thresholds list 객체 내의 값을 iteration 하면서 평가 수행
    for custom_threshold in thresholds:
        binarizer = Binarizer(threshold=custom_threshold).fit(pred_proba_c1)
        custom_predict = binarizer.transform(pred_proba_c1)
        print('\n임계값: ', custom_threshold)
        get_clf_eval(y_test, custom_predict)
```

```python
# split 
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, roc_auc_score
from sklearn.metrics import f1_score, confusion_matrix, precision_recall_curve, roc_curve
from sklearn.preprocessing import StandardScaler, Binarizer
from sklearn.ensemble import RandomForestClassifier

# 모델 변경 및 grid search 적용
# penalty = ['l1', 'l2'] # 페널티(penalty) 하이퍼파라미터 값의 후보를 만듭니다.
# C = np.logspace(0, 4, 10) # 규제 하이퍼파라미터 값의 후보 범위를 만듭니다.
# hyperparameters = dict(C=C, penalty=penalty) # 하이퍼파라미터 후보 딕셔너리를 만듭니다.

# lr= LogisticRegression(random_state=0,max_iter=1000)
# lr_clt = GridSearchCV(lr, hyperparameters, cv = 5,verbose=0)
# lr_clt = RandomForestClassifier(n_estimators=5)

params = { 'n_estimators' : [10, 100],
           'max_depth' : [6, 8, 10, 12],
            }

# # RandomForestClassifier 객체 생성 후 GridSearchCV 수행
rf_clf = RandomForestClassifier()
lr_clt = GridSearchCV(rf_clf, param_grid = params, cv = 3)

# StandardScaler 클래스를 상용하여 데이터 세트에 스케일링 적용
# 기존에 가지고 있는 테스트 set
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_train)

# test samle이 없으므로 임의로 나눔
# 기존 X_sclaed 값을 --> X_train / X_text로 나눔
X_scaled_train, X_scaled_test, y_sacled_train, y_scaled_test = train_test_split(X_scaled, y_train, test_size = 0.2)

# 학습해서 결과 계속 확인할 것 
lr_clt.fit(X_scaled_train, y_sacled_train)
pred = lr_clt.predict(X_scaled_test)
accuracy = accuracy_score(y_scaled_test, pred)
print('랜덤 포레스트 정확도: {:.4f}'.format(accuracy))

print("---------------")

# # 임계값 변경
# thresholds = [0.3, 0.33, 0.36, 0.39, 0.42, 0.45, 0.48, 0.50]
# pred_proba = lr_clt.predict_proba(X_scaled_test)
# get_eval_by_threshold(y_scaled_test, pred_proba[:, 1].reshape(-1, 1), thresholds)
```
